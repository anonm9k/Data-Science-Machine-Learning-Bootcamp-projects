```python
# imports
import sklearn 
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier

import pandas as pd
import numpy as np

from sklearn import linear_model, preprocessing

# data
data = pd.read_csv('car.data')

# preprocessing: transform all data to numbers
le = preprocessing.LabelEncoder() # used to encode 'list' of data

buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))

# label
predict = "class"

# create features and label
X = list(zip(buying, maint, door, persons, lug_boot, safety, cls)) # zip will create tuples and merge all features into a list
y = list(cls)

# splitting data
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1)

# using knn algorithm
model = KNeighborsClassifier(n_neighbors=5) # how many neighbors
model.fit(X_train, y_train)

acc = model.score(X_test, y_test)
acc
```

```python
predicted = model.predict(X_test) # list of predicted values
names = ['unacc', 'acc', 'good', 'vgood']

for n in range(len(predicted)):
    print('predicted: ', names[predicted[n]], 'Actual: ', names[y_test[n]])
```
