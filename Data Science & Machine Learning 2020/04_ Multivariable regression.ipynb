### Steps before running multiple regression
1. Generating potential variables; independent and dependent.
2. Collect data on variables.
3. Check relationships between each independent variables and dependent variable using correlations and scatterplot.
4. Check relationship among independent variables themselves. (Checking for multicolinearity)
5. (Optional) Conduct simple linear regression for each IV/DV pair.
6. Use non-redundant Independent variables to find best fitting model.
7. Use best fitting model to predict the dependent variable

### The main idea of a linear regression
1. Use least squares to fit a line to the data.
2. Calculate $R^2$.
3. Calculate p Value for $R^2$

### $$Variance = \frac{Data - Line}{Datas(n)}$$
Variation around mean line is: $$Var(Mean) = \frac{Data - Mean}{n}$$
Variation around fit line(our regression line) is: $$Var(Fit) = \frac{Data - Fit}{n}$$


$R^2$ tells us how much of the variation of Data($y$) is explained by the line $x$

$$R^2 = \frac{Var(Mean)- Var(Fit)}{Var(Mean)}$$

### Multicollinearity: 
When the correlation between two variable is very high. Which can be a problem so it must be investigated. For example: high body fat % is highly correlated with weight. But individuals can have very low fat % and also have high weight like bodybuilders. So each of the features do not provide unique and independent information to the regression.

### Discrete vs Continues variables:
Discrete variable can only take certain values (e.g 0 or 1). Continues variables can take any value in a particuler limit (e.g. from 1 to 100). 
Discrete variables are better to be excluded from the dataset as irrelevent data. As correlation calculations only assume continues values.

## Training & Test Dataset Split

```python
prices = data['PRICE']
features = data.drop('PRICE', axis=1) #create a new datafram excluding the price column(axis1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state= 10)
X_train
```

## Multivariate Regression
## $$PR\hat{I}CE = \theta_0 + \theta_1 RM + \theta_2 NOX + ... + \theta_{13} LSTAT $$
Each $\theta_x$ is the coef. And $\theta_0$ is the intercept.

```python
regr = LinearRegression()
regr.fit(X_train, y_train)


print('Intercept 𝜃0: ', regr.intercept_) #All theta values
print('training data R-squared values: ',regr.score(X_train, y_train) ) #How well our training model fit the data
print('Testing data R-squared values: ', regr.score(X_test, y_test)) #How well our testing model fit the data

#coefs of training data
pd.DataFrame(data=regr.coef_, index=X_train.columns, columns=['Coef 𝜃x'])
```

```python
#coefs of testing data
regr = LinearRegression()
regr.fit(X_test, y_test)

pd.DataFrame(data=regr.coef_, index=X_test.columns, columns=['Coef 𝜃x'])
```

## Testing for Multicolinearity
$$TAX = \alpha_0 + \alpha_1 RM + \alpha_2 NOX +...+ \alpha_{12} LSTAT$$
$$ VIF_{TAX} = \frac{1}{1 - R_{TAX} ^ 2} $$

```python
variance_inflation_factor(exog=x_incl_const.values, exog_idx=1) #gets vif for CRIM(exog_idx=1)
```

```python
# finding all the VIF's using loop

for n in range(x_incl_const.shape[1]):
    print(variance_inflation_factor(exog=x_incl_const.values, exog_idx=n))
```

```python
vif = [] #empty list
for n in range(x_incl_const.shape[1]):
    vif.append(variance_inflation_factor(exog=x_incl_const.values, exog_idx=n))
print(vif)

pd.DataFrame({'coef_name': x_incl_const.columns, 'vif': np.around(vif, 2)})
```

## Residuals & Residual Plots
Why normal distribution of residuals matter?
To be able to calculate the range from standard deviation of normal distribution

### Modified Model: using log prices

```python
# Modified model: transformed (using log prices) & simplified (dropping two features age and indus)
prices = np.log(data['PRICE']) # Use log prices
features = data.drop(['PRICE', 'INDUS', 'AGE'], axis=1) #feature without indus and age

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, random_state=10)

# Using Statsmodel
X_incl_const = sm.add_constant(X_train) #X_train with constant
model = sm.OLS(y_train, X_incl_const) #making model
results = model.fit() #fitting regression

# Residuals
# residuals = y_train - results.fittedvalues
# results.resid

# Graph of Actual vs. Predicted Log Prices
corr = round(y_train.corr(results.fittedvalues), 2)

plt.scatter(x=y_train, y=results.fittedvalues, c='navy', alpha=0.6)
plt.plot(y_train, y_train, color='cyan')

plt.xlabel('Actual log prices $y _i$', fontsize=14)
plt.ylabel('Prediced log prices $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted log prices: $y _i$ vs $\hat y_i$ (Corr {corr})', fontsize=17)

plt.show()

#Graph of Actual vs Predicted Prices(reversing log)
plt.scatter(x=np.e**y_train, y=np.e**results.fittedvalues, c='blue', alpha=0.6)
plt.plot(np.e**y_train, np.e**y_train, color='cyan')

plt.xlabel('Actual prices 000s $y _i$', fontsize=14)
plt.ylabel('Prediced prices 000s $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted prices: $y _i$ vs $\hat y_i$ (Corr {corr})', fontsize=17)

plt.show()

# Residuals vs Predicted values

plt.scatter(x=results.fittedvalues, y=results.resid, c='navy', alpha=0.6)

plt.xlabel('Predicted log prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values', fontsize=17)

plt.show()

# Mean Squared Error & R-Squared
reduced_log_mse = round(results.mse_resid, 3)
reduced_log_rsquared = round(results.rsquared, 3)
```

```python
# Distribution of Residuals (log prices) - checking for normality
#Skew is zero for normla distribution

resid_mean = round(results.resid.mean(), 3)
resid_skew = round(results.resid.skew(), 3)

sns.distplot(results.resid, color='navy')
plt.title(f'Log price model: residuals Skew ({resid_skew}) Mean ({resid_mean})')
plt.show()
```

### Original Model

```python
# Original model: normal prices & all features
prices = data['PRICE']
features = data.drop(['PRICE'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, random_state=10)

X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# Graph of Actual vs. Predicted Prices
corr = round(y_train.corr(results.fittedvalues), 2)
plt.scatter(x=y_train, y=results.fittedvalues, c='indigo', alpha=0.6)
plt.plot(y_train, y_train, color='cyan')

plt.xlabel('Actual prices 000s $y _i$', fontsize=14)
plt.ylabel('Prediced prices 000s $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted prices: $y _i$ vs $\hat y_i$ (Corr {corr})', fontsize=17)

plt.show()

# Residuals vs Predicted values
plt.scatter(x=results.fittedvalues, y=results.resid, c='indigo', alpha=0.6)

plt.xlabel('Predicted prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values', fontsize=17)

plt.show()

# Residual Distribution Chart
resid_mean = round(results.resid.mean(), 3)
resid_skew = round(results.resid.skew(), 3)

sns.distplot(results.resid, color='indigo')
plt.title(f'Residuals Skew ({resid_skew}) Mean ({resid_mean})')
plt.show()

# Mean Squared Error & R-Squared
full_normal_mse = round(results.mse_resid, 3)
full_normal_rsquared = round(results.rsquared, 3)
```

### Modified Model: ommiting key features and using log prices

```python
# Model Omitting Key Features using log prices
prices = np.log(data['PRICE'])
features = data.drop(['PRICE', 'INDUS', 'AGE', 'LSTAT', 'RM', 'NOX', 'CRIM'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, random_state=10)

X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# Graph of Actual vs. Predicted Prices
corr = round(y_train.corr(results.fittedvalues), 2)
plt.scatter(x=y_train, y=results.fittedvalues, c='#e74c3c', alpha=0.6)
plt.plot(y_train, y_train, color='cyan')

plt.xlabel('Actual log prices $y _i$', fontsize=14)
plt.ylabel('Predicted log prices $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted prices with omitted variables: $y _i$ vs $\hat y_i$ (Corr {corr})', fontsize=17)

plt.show()

# Residuals vs Predicted values
plt.scatter(x=results.fittedvalues, y=results.resid, c='#e74c3c', alpha=0.6)

plt.xlabel('Predicted prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values', fontsize=17)

plt.show()

# Mean Squared Error & R-Squared
omitted_var_mse = round(results.mse_resid, 3)
omitted_var_rsquared = round(results.rsquared, 3)
```

```python
pd.DataFrame({'R-Squared': [reduced_log_rsquared, full_normal_rsquared, omitted_var_rsquared],
             'MSE': [reduced_log_mse, full_normal_mse, omitted_var_mse], 
             'RMSE': np.sqrt([reduced_log_mse, full_normal_mse, omitted_var_mse])}, 
            index=['Reduced Log Model', 'Full Normal Price Model', 'Omitted Var Model'])
```

#### Range is between two standard deviations($\sigma$) . A $\sigma$ is Rooted MSE(RMSE)

```python
# Challenge: Our estimate for a house price is $30,000. Calculate the upper and lower bound
# for a 95% prediction interval using the reduced log model

print('1 s.d. in log prices is', np.sqrt(reduced_log_mse))
print('2 s.d. in log prices is', 2*np.sqrt(reduced_log_mse))

upper_bound = np.log(30) + 2*np.sqrt(reduced_log_mse)
print('The upper bound in log prices for a 95% prediction interval is ', upper_bound)
print('The upper bound in normal prices is $', np.e**upper_bound * 1000)

lower_bound = np.log(30) - 2*np.sqrt(reduced_log_mse)
print('The lower bound in log prices for a 95% prediction interval is ', lower_bound)
print('The lower bound in normal prices is $', np.e**lower_bound * 1000)
```
