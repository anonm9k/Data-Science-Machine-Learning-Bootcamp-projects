# Imports

````python
import numpy as np
import tensorflow as tf

from tensorflow.compat.v1.saved_model import load, tag_constants
````

# Constants

````python
X_TEST_PATH = 'MNIST/load_xtest.csv'
Y_TEST_PATH = 'MNIST/load_ytest.csv'
````

# Load the Data

````python
y_test = np.loadtxt(Y_TEST_PATH, delimiter=',', dtype=int)
print(f'Shape of y_test {y_test.shape}')
````

````python
x_test = np.loadtxt(X_TEST_PATH, delimiter=',', dtype=float)
print(f'Shape of x_test {x_test.shape}')
````

````python
np.savetxt(fname='x_test0_ylabel7.txt', X=[x_test[0]], delimiter=', ', fmt='%1.2f')
````

# Create Session

````python
graph = tf.compat.v1.Graph()
sess = tf.compat.v1.Session(graph = graph) # link graph with session
````

# Load the Model

```
tf.compat.v1.saved_model.load(sess,
                           [tag_constants.SERVING],
                           EXPORT_DIR)
```

````python
# loading saved model
load(sess=sess, tags=[tag_constants.SERVING], export_dir='SavedModel')
````

# Run Session & Make Prediction

````python
# make variable to store inputs(X)
X = graph.get_tensor_by_name('X:0') # X is the operation, adding 0 to take hold of the value

# make variable to store prediction
y_pred = graph.get_tensor_by_name('accuracy_calc/Prediction:0')
````

````python
# start session to run the numbers and get prediction result
prediction = sess.run(fetches= y_pred, feed_dict={X: x_test})
````

# Check Results

````python
y_test[:5]
````

````python
prediction[:5]
````

````python
nr_correct = sum(np.equal(prediction, y_test))
nr_correct / len(prediction) * 100
````

# Cleanup

````python
sess.close()
tf.compat.v1.reset_default_graph()
````
